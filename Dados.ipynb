```python
import random

def lanzar_dados(): #función para lanzar 2 dados
  while True:
    try:
      numd = int(input("Lanzar 1 o 2 dados: "))
      if numd in [1, 2]:
        break
      else:
        print("Ingresar 1 o 2.")
    except ValueError:
      print("Ingresar un número.")

  resultados = []
  for _ in range(numd):
    resultados.append(random.randint(1, 6))

  if numd == 1:
    print(f"Dado 1: {resultados[0]}")
  else:
    print(f"Dado 1: {resultados[0]} , dado 2: {resultados[1]}.")

lanzar_dados()
```
